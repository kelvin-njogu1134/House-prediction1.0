```python
#sklearn modules

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#pandas library
import pandas as pd
```

```python
#loading csv 
df = pd.read_csv("salary.csv")

#defining features(x) and target (y)
X = df[["YearsExperience","Age","EducationLevel","HoursPerWeek"]]
y = df["Salary"]
#spliting the x train,test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.200, random_state = 42)
#create module
model=LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print(f"predictions:{y_pred}")
print(f"actual data:{y_test.values}")

import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred, color = "red")
plt.title("years,age,educationlevel,hour/week VS Salary")
plt.xlabel("Actual")
plt.ylabel("predictons")
plt.plot([y_test.min(),y_test.max()],
         [y_test.min(),y_test.max()]
         
        )

plt.show()




```
