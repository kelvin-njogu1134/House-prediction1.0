```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.jumia.co.ke/mlp-black-friday-h-entertainment-center/televisions/?display_size=43--43.0"
```

```python
import requests
```

```python
response =requests.get(url)
response
```

```python
htmlcode = BeautifulSoup(response.text,"html.parser")
#htmlcode
```

```python
itemname  = htmlcode.find_all(class_="name")
itemname
```

```python
itemlist=[]
for item in itemname:
    itemlist.append(item.text)

itemlist
    
```

```python
pricedata = htmlcode.find_all(class_="prc")


    
```

```python
pricelist=[]
for item in pricedata:
    pricelist.append(item.text)
pricelist
```

```python
import pandas as pd

df = pd.DataFrame({
        "price":pricelist,
        "product":itemlist}
                 )

df
```
