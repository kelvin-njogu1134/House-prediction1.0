```python
import requests
```

```python
endpoint = "https://api.worldbank.org/v2/country/KEN;NG;ZA;IN/indicator/NY.GDP.MKTP.KD.ZG?format=json&date=2000:2023"

info = requests.get(endpoint)

data = info.json()
```

```python
import pandas as pd


data_list = data[1]

records = []

for item in data_list:
    records.append({
        "year": int(item["date"]),               
        "country": item["country"]["value"],     
        "value": int(item["value"]),
        "ISO":item["countryiso3code"],
        "indicator": item["indicator"]["value"]
    })


df = pd.DataFrame(records)


df = df.sort_values("year", ascending=False).reset_index(drop=True)

print(df)
```

```python
country_total = df.groupby("country")['value'].sum()

print(country_total)
```

```python
best_peryear = df.loc[df.groupby("year")["value"].idxmax()]
best_peryear
```
