```python
url = "https://www.airbnb.com/s/Kilimani--Nairobi/homes?refinement_paths%5B%5D=%2Fhomes&place_id=ChIJVfqQ_Q0aLxgRfLwhb3oF9Tc&acp_id=908bc058-de0c-4130-a0a3-59abf2f94832&date_picker_type=calendar&search_type=user_map_move&query=Kilimani%2C%20Nairobi&flexible_trip_lengths%5B%5D=one_week&monthly_start_date=2026-05-01&monthly_length=3&monthly_end_date=2026-08-01&search_mode=regular_search&price_filter_input_type=2&channel=EXPLORE&ne_lat=-1.2794257752938292&ne_lng=36.79351407648949&sw_lat=-1.2954775237891827&sw_lng=36.78071880786757&zoom=15.157670957602754&zoom_level=15.157670957602754&search_by_map=true&price_filter_num_nights=5"
```

```python
import requests
```

```python
response = requests.get(url)
```

```python
htmlcode = BeautifulSoup(response.text,"html.parser")

htmlcode
```

```python

airbnbprice = htmlcode.find_all(class_="u174bpcy atm_7l_mhi6a6 atm_cs_80qqwn atm_rd_14k51in atm_rq_glywfm atm_cs_l3jtxx__1v156lz dir dir-ltr")
```

```python
pricelist=[]
for item in airbnbprice:
    pricelist.append(item.text)
pricelist
    
```

```python
clean_prices = []

for price in pricelist:
    p = price.replace("\xa0", "") \
             .strip()
    
    clean_prices.append(p)

clean_prices
```

```python
airbnbname = htmlcode.find_all(class_="t1jojoys atm_g3_1kw7nm4 atm_ks_15vqwwr atm_sq_1l2sidv atm_ld_l4hv9n atm_gp_7iojc0 atm_9s_cj1kg8 atm_6w_1e54zos atm_fy_1vgr820 atm_7l_hfv0h6 atm_cs_1mexzig atm_w4_1eetg7c atm_ks_zryt35__1rgatj2 dir dir-ltr")
airbnbname
```

```python
namelist = []
for name in airbnbname:
    namelist.append(name.text)

namelist
    
```

```python
import pandas as pd

df = pd.DataFrame({
    "Location":namelist,
    "price":clean_prices})
```
