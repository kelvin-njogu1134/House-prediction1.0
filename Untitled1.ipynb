```python
endpoint = "https://v6.exchangerate-api.com/v6/52bb380eef234f8a36821a15/latest/USD"
```

```python
data = requests.get(endpoint)
```

```python
currencydata = data.json()
```

```python
currencydata
```

```python
rates = currencydata["conversion_rates"]
```

```python
#convertion

def convert(amount,from_currency,to_currency):
    if from_currency != currencydata['base_code']:
        amount_inbase = amount/rates[from_currency]
    else:
        amount_inbase = amount

    converted_amount = amount_inbase * rates[to_currency]
    return converted_amount
    
results = convert(1000,'KES','USD')

print(f"1000 kes is {results} usd")
 


    
```

```python



import requests
import pandas as pd

indicator = "NY.GDP.MKTP.KD.ZG"
base_url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator}"

all_records = []
page = 1

while True:
    url = f"{base_url}?format=json&date=2000:2023&page={page}"
    response = requests.get(url)
    data = response.json()

    # Stop when no more data
    if len(data) < 2:
        break

    for item in data[1]:
        # ✅ Keep ONLY real countries
        if item["countryiso3code"]:  # not empty
            all_records.append({
                "year": int(item["date"]),
                "country": item["country"]["value"],
                "country_iso3": item["countryiso3code"],
                "value": item["value"]
            })

    page += 1

# Create DataFrame
df = pd.DataFrame(all_records)

# Clean missing values
df = df.dropna()

# Sort nicely
df = df.sort_values(["country", "year"]).reset_index(drop=True)

df
```
