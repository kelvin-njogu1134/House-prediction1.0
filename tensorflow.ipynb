```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
#loading and spliting
(train_images,train_label),(test_images,test_label) = keras.datasets.fashion_mnist.load_data()
```

```python
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
#model
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(128,activation ="relu"),
    keras.layers.Dense(10,activation = "softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(train_images,train_label, epochs = 10)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_label)
print("Accuracy:",test_acc)
```

```python
prediction = model.predict(test_images)
index = 1
predicted_label = np.argmax(prediction[index])
print("predictions:", class_names[predicted_label])
print("actual:", class_names[test_label[index]])

plt.imshow(test_images[index], cmap = "gray")
plt.title(class_names[predicted_label])
plt.show()
```
